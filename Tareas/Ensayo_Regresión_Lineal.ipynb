# 2. Historia

Los primeros aportes que se realizaron y de los cuales surgió la Regresión Lineal estuvieron asociados a la Correlación entre variables en problemas de Biología, Biometría y Eugenesia. 

El primer autor que se interesó en el tema fue Lambert-Adolphe-Jacques Quetelet (1796-1874), sus aportes sobre la correlación y regresión se originaron en sus estudios sobre el hombre medio, estimando empíricamente las medias y desviaciones típicas de medidas antropométricas, las cuales suponía que dependen de varias variables independientes tales como el sexo, edad, profesión o nivel de educación. Estepa, A. et al. (2012)



Otro personaje que se interesó por el tema fue  Francis Galton (1822-1911) estudió Medicina y Matemáticas, sin embargo, no era conocedor de los métodos estadísticos de la época (usados, por ejemplo, por Laplace y Gauss). Por medio de investigaciones empíricas, estudió la variabilidad de características humanas. Su inspiración fue el evidenciar que los hijos de padres muy altos tendían a ser más bajos que sus padres y los hijos de padres muy bajos tendían a ser más altos que sus padres. Era como si una fuerza misteriosa estuviera provocando que las alturas humanas se alejaran de los extremos y se acercaran a la media o media de todos los humanos. Galton llamó a este fenómeno "*regresión a la media*". Salsburg, D. (2002)

Pearson continúa el trabajo de Galton, interesado en la Biometría debido a los trabajos de éste y de Weldon, quien en 1892 fue el primero que publicó un artículo en el que señala el significado de un coeficiente de correlación negativo.

Las ideas modernas sobre regresión se originan en los trabajo de Legendre y Gauss, sobre el método de mínimos cuadrados, para ajustar los datos sobre las órbitas de cuerpos celestes. 

En 1829 Gauss fue capaz de establecer la razón por la cual este procedimiento es muy adecuado desde el punto de vista estadístico, mediante lo que hoy se conoce como teorema de Gauss-Markov, que muestra que los estimadores obtenidos con este método son insesgados y no se requiere una distribución específica para los datos que se ajustan. El impulso posterior lo dan los trabajos citados de Galton y Pearson sobre la herencia. Yule la utiliza para el estudio de fenómenos sociales, como las causas de pobreza alrededor de 1899. 

Alrededor de 1895, Pearson había resuelto las propiedades matemáticas del coeficiente de correlación y la regresión simple utilizada para la predicción lineal entre dos variables continuas. También generaliza la idea de regresión de Galton a varias dimensiones, comprobando que muchas variables biométricas siguen la distribución normal multivariante y por tanto se podría aplicar la regresión múltiple para relacionar varias de estas medidas. # 3. Algoritmo

Esta sección se basa en James, G. et al. (2013)

Bajo la regresión lineal simple, se pretende obtener una respuesta cuantitativa $Y$ sobre la base de una única variable predictora de regresión $X$. Se supone que hay aproximadamente una relación lineal entre $X$ y $Y$. Lo anterior, se puede escribir matemáticamente de la siguiente forma: 

$$Y= \beta_{0} + \beta_{1} X + \epsilon \text{, con } E(\epsilon)=0 \text{  y  } V(\epsilon)=\sigma^{2}$$

donde, $Y$ es la variable respuesta dependiente, $X$ la variable regresora independiente, $\beta_{0}$ y $\beta_{1}$ parámetros o coeficientes del modelo (constantes desconocidas) y $\epsilon$ vector de errores.

De tal forma que al estimar los parámetros, se puede obtener $\hat{y}$ el cual indica una predicción de $Y$ sobre la base de $X = x$.

$$\hat{y} = \hat{\beta}_{0} + \hat{\beta}_{1} x$$

Pero en la práctica $\beta_{0}$ y $\beta_{1}$ son desconocidos, así que ¿de qué forma se obtinen las estimaciones para estos parámetros?

## 3.1 Estimación

Sean $(x_{1},y_{1}),\text{ } (x_{2},y_{2}),\text{ } ...,\text{ } (x_{n},y_{n})$,  $n$ pares de observaciones, cada uno de los cuales consta de una medida de $X$ y una medida de $Y$.

El objetivo es obtener estimaciones de coeficientes $\hat{\beta}_{0}$ y $\hat{\beta}_{1}$ de manera que el modelo lineal presentado, se ajuste bien a los datos disponibles. En otras palabras, encontrar una intersección $\hat{\beta}_{0}$ y una pendiente  $\hat{\beta}_{1}$ tal que la línea resultante esté lo más cerca posible de los $n$ puntos de datos.



# 3.1.1 Estimación Mínimos Cuadrados




El ajuste se encuentra minimizando la suma de errores cuadrados. La expresión $e_{i}=y_{i}-\hat{y}_{i}$ representa el i-ésimo residuo: esta es la diferencia entre el i-ésimo valor de respuesta observado y el i-ésimo valor de respuesta que predice nuestro modelo lineal.

En la siguiente imagen, cada segmento de línea gris representa un error y el ajuste se logra al promediar sus cuadrados. 

min_cuadr.PNG 

Tomado de James, G. et al. (2013)

Se define la suma de cuadrados residual (RSS) como: $RSS = e^{2}_{1} + e^{2}_{2}
+ · · · + e^{2}_{n}$. El enfoque de mínimos cuadrados elige $\hat{\beta}_{0}$ y $\hat{\beta}_{1}$ tal que se minimice el RSS. Usando algo de cálculo, se puede demostrar que los minimizadores son:
$$\hat{\beta}_{1} = \dfrac{\sum_{i=1}^{n}(x_{i}-\bar{x})(y_{i}-\bar{y})}{\sum_{i=1}^{n}(x_{i}-\bar{x})^{2}}$$

$$\hat{\beta}_{0} = \bar{y} - \hat{\beta}_{1}\bar{x}$$

donde $\bar{y} = \dfrac{1}{n} \sum_{i=1}^{n}y_{i}$ y $\bar{x} = \dfrac{1}{n} \sum_{i=1}^{n}x_{i}$ son las medias muestrales. 

En el ejemplo que se presenta en la siguiente sección se verá gráficamente la optimización realizada para encontrar $\hat{\beta}_{0}$ y $\hat{\beta}_{1}$ a través de los gráficos de contorno y tridimensionales del RSS.

Posterior al ajuste del modelo, es necesario evaluar la precisión de las estimaciones de coeficientes y determinar si la relación existente entre las dos variables puede ser descrita de forma adecuada por el modelo. La verdadera relación probablemente no sea lineal o pueden existir otras variables que causen variación en $Y$.

Además, se debe evaluar la precisión del modelo, pues es natural querer cuantificar la medida en que el modelo se ajusta a los datos. La calidad de un ajuste de regresión lineal generalmente se evalúa utilizando dos cantidades relacionadas: el error estándar residual $(RSE)$ y la estadística $R^{2}$.

La regresión lineal simple es un enfoque útil para predecir una respuesta sobre la base de una única variable predictora. Sin embargo, en la práctica, a menudo tenemos más de un predictor. En este caso, se extienden los resultados de la Regresión Lineal Simple de tal forma que pueda considerar los múltiples predictores, a esta metodología se le conoce como Regresión Lineal Múltiple.

Cuando ajustamos un modelo de regresión lineal a un conjunto de datos en particular, muchos problemas pueden ocurrir. Los más comunes entre estos son los siguientes:
1. No linealidad de las relaciones respuesta-predictor.
2. Correlación de términos de error. Un supuesto importante del modelo de regresión lineal es que los términos de error, $e_{1}, e_{2}, · · ·, e_{n}$, no están correlacionados.
3. Varianza no constante de los términos de error. El supuesto del modelo de regresión lineal es que los términos de error tienen una varianza constante. Los errores estándar, los intervalos de confianza y las pruebas de hipótesis asociados con el modelo lineal se basan en este supuesto.
4. Valores atípicos.
5. Colinealidad. La colinealidad se refiere a la situación en la que dos o más variables predictoras están estrechamente relacionadas entre sí.

En la práctica, identificar y superar estos problemas es tanto un
el arte como ciencia.

# 4. Ejemplo

Continuando con la situación presentada en la introducción, se tiene un conjunto de datos publicitarios con 200 registros y cuatro variables: 

*   Sales: ventas (en miles de unidades) de un producto en particular
*   TV: Presupuestos publicitarios (en miles de dólares) para televisión
*   Radio: Presupuestos publicitarios (en miles de dólares) para radio
*   Newspaper: Presupuestos publicitarios (en miles de dólares) para periódicos

Consiste en las ventas de un producto en particular en 200 mercados diferentes, junto con los presupuestos publicitarios del producto en cada uno de esos mercados para tres medios de comunicación diferentes: televisión, radio y periódicos.

El objetivo es identificar de qué forma los presupuestos en cada uno de los medios de comunicación pueden incrementar las ventas del producto.

Este ejemplo está basado en el código asociado al libro An Introduction to Statistical Learning with Applications in R, James, G. et al. (2013).

```
# Cargar las librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
import seaborn as sns

from sklearn.preprocessing import scale
import sklearn.linear_model as skl_lm
from sklearn.metrics import mean_squared_error, r2_score
import statsmodels.api as sm
import statsmodels.formula.api as smf

%matplotlib inline
plt.style.use('seaborn-white')
```

```
# Cargar la base de datos
advertising = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/UN_Mineria_de_datos/Tareas/Ensayo Reg Lineal/Advertising.csv', usecols=[1,2,3,4])
advertising.info()
```

En el siguiente gráfico se presenta la distribución de las ventas en función del presupuesto publicitario para televisión y se ajusta una recta al comportamiento. Sin embargo, más adelante se verá la estimación de los parámetros.

```
sns.regplot(advertising.TV, advertising.Sales, order=1, ci=None, scatter_kws={'color':'r', 's':9})
plt.xlim(-10,310)
plt.ylim(ymin=0);
```

A continuación se realiza el ajuste y se presentan las estimaciones $\hat{\beta}_{0}$ y $\hat{\beta}_{1}$ obtenidas bajo la optimización presentada en la anterior sección. 

```
# Coeficientes de Regresión (Mínimos Cuadrados Ordinarios)
regr = skl_lm.LinearRegression()

# Se centra la variable TV y se definen variables
#X = scale(advertising.TV, with_mean=True, with_std=False).reshape(-1,1)
# Sin centrar la variable TV y se definen variables
X = scale(advertising.TV, with_mean=False, with_std=False).reshape(-1,1)
y = advertising.Sales

# Ajuste de la regresión lineal
regr.fit(X,y)
print("La estimación para beta_0 es: ",regr.intercept_)
print("La estimación para beta_1 es: ",regr.coef_)
```

La función cuadrática objeto de la optimización, se puede ver representada a continuación. De esta forma se ve gráficamente el punto $(\beta_{0},\beta_{1})$ que hace mínima la expresión RSS.

```
# Crea las coordenadas de la grilla para graficar
B0 = np.linspace(regr.intercept_-2, regr.intercept_+2, 50) # se generan 50 números comprendidos entre +- 2 alrededor de beta_0
B1 = np.linspace(regr.coef_-0.02, regr.coef_+0.02, 50) # se generan 50 números comprendidos entre +- 0.02 alrededor de beta_1
xx, yy = np.meshgrid(B0, B1, indexing='xy') # retorna las coordenadas matriciales en x y en y
Z = np.zeros((B0.size,B1.size)) # crea matriz de 0 de tamaño 50x50

# Calcula los valores Z (RSS) basado en la grilla construida
for (i,j),v in np.ndenumerate(Z):
    Z[i,j] =((y - (xx[i,j]+X.ravel()*yy[i,j]))**2).sum()/1000

# Minimized RSS
min_RSS = r'$\beta_0$, $\beta_1$ for minimized RSS'
min_rss = np.sum((regr.intercept_+regr.coef_*X - y.values.reshape(-1,1))**2)/1000
print("El valor mínimo para esta función es ",min_rss)
```

```
fig = plt.figure(figsize=(15,6))
fig.suptitle('RSS - Coefficients de Regresión', fontsize=20)

# crea los esqueletos para los gráficos
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122, projection='3d')

# Gráfico de contorno
CS = ax1.contour(xx, yy, Z, cmap=plt.cm.Set1, levels=[2.15, 2.2, 2.3, 2.5, 3])
ax1.scatter(regr.intercept_, regr.coef_[0], c='r', label=min_RSS)
ax1.clabel(CS, inline=True, fontsize=10, fmt='%1.1f')

# Gráfico tridimensional
ax2.plot_surface(xx, yy, Z, rstride=3, cstride=3, alpha=0.3)
ax2.contour(xx, yy, Z, zdir='z', offset=Z.min(), cmap=plt.cm.Set1,
            alpha=0.4, levels=[2.15, 2.2, 2.3, 2.5, 3])
ax2.scatter3D(regr.intercept_, regr.coef_[0], min_rss, c='r', label=min_RSS)
ax2.set_zlabel('RSS')
ax2.set_zlim(Z.min(),Z.max())
ax2.set_ylim(0.02,0.07)

# ajuste para los dos gráficos
for ax in fig.axes:
    ax.set_xlabel(r'$\beta_0$', fontsize=17)
    ax.set_ylabel(r'$\beta_1$', fontsize=17)
    ax.set_yticks([0.03,0.04,0.05,0.06])
    ax.legend()
```

En la siguiente tabla se presentan las estimaciones junto con algunas estadísticas, de aquí se puede concluir que la variable de Presupuesto publicitario para televisión resulta ser significativa en la estimación de las ventas del producto. 

Según las estimaciones, $ 1,000 adicionales gastados en publicidad televisiva están asociados con la venta de aproximadamente 47.5 unidades adicionales del producto.
